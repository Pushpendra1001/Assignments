```python
Q.1 Explain why we have to use the Exception class while creating a custom Exception.
Ans We use the Exception class as the base class for creating custom exceptions because 
     it provides a set of standard behaviors and attributes that are essential for 
     consistent and reliable error handling. By inheriting from the Exception class, 
     our custom exception will automatically inherit these behaviors and attributes, such 
     as the ability to print an error message, display a traceback, and raise the exception 
     in a standardized way.
```

```python
#Q.2 Write a python program to print python Exception hierarchy. 
def hierarchy(exceptions, k=0):
    for i in exceptions:
        print(' ' * k, i.__name__)
        hierarchy(i.__subclasses__(), k + 1)
hierarchy([Exception])
```

```python
Q.6 List down the best practices for Exception hadling in python .
Ans : Exception handling is an important part of any Python program, as it helps to handle
    runtime errors gracefully and make the program more robust. Here are some best practices
    for exception handling in Python:
    
    1. Catch specific exceptions: Instead of catching all exceptions with a generic except
    block, it is better to catch specific exceptions using multiple except blocks. This makes
    the code more readable and easier to debug.
    
    2. Use finally block: The finally block is used to execute cleanup code, such as closing
    files or database connections, regardless of whether an exception occurred or not. This 
    ensures that resources are released properly and the program does not leave any open handles.
    
    3. Don't ignore exceptions: It is a bad practice to catch an exception and do nothing with
    it. At a minimum, log the exception with an appropriate message to help with debugging.
    
    4. Don't use exceptions for flow control: Exceptions should not be used to control 
    the flow of the program, as they are designed for handling errors. Use control structures 
    like if, else, and while for flow control.
    
    5. Keep the try block small: The try block should contain only the code that may raise
    an exception. Keeping the try block small makes it easier to locate and fix errors.
    
    6. Use custom exception classes: In some cases, it may be useful to create custom exception 
    classes for specific error conditions in the program. This can make the code more readable
    and easier to maintain.
    
    7. Use meaningful exception messages: Exception messages should be meaningful and informative, 
    so that they can help with debugging. They should explain what went wrong and provide enough 
    information to help locate the problem.
```
