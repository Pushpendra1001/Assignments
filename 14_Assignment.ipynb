```python
Q.1  What is multithreading in python? Why is it used? Name the module used to handle threads in python.
Ans: Multithreading in Python refers to the process of executing multiple threads or tasks simultaneously
    within a single program. Each thread runs independently of the others, allowing for more efficient use of 
    system resources and faster execution of tasks.

    Multithreading is used in Python to improve the performance of programs that involve I/O operations,
    such as web scraping, downloading files, or processing large data sets. By running multiple threads 
    simultaneously, it is possible to reduce the overall execution time of a program.

    In Python, the threading module is used to handle threads. This module provides a simple way to 
    create and manage threads within a Python program. It includes functions for starting, stopping, 
    and managing threads, as well as synchronization primitives like locks and semaphores that allow 
    threads to communicate with each other and coordinate their actions.
```

```python
 Q.2 Why threading module used? Write the use of the following functions:
        activeCount()
        currentThread()
         enumerate()
Ans  The threading module in Python is used for creating, managing, and synchronizing multiple threads 
      of execution within a single process. It provides a simple and efficient way to write concurrent code in Python.
    
      activeCount(): This function returns the number of thread objects that are active in the current Python interpreter. 
      An active thread is a thread that has been started but has not yet finished or been joined.
    
      currentThread(): This function returns a reference to the current thread object that is executing the current Python code. 
      This function can be used to get information about the current thread, such as its name, ID, and other attributes

      enumerate(): This function returns a list of all thread objects that are active in the current Python interpreter. 
      Each item in the list is a reference to a thread object, and the list is ordered by the order in which the threads 
      were started.
```

```python
Q.3 Explain the following functions in  multithreading
     run()
     start()
     join()
     isAlive()
Ans: run(): This method is called when a thread is started using the start() method. It is the 
     method where the thread's activity is defined. By default, it does nothing, but you can override 
     it in your own class to define a thread's behavior.

     start(): This method starts a thread's activity by calling the run() method. When you call start(), 
     a new thread is created, and the run() method of that thread is executed in parallel with the main thread. 
     Once the start() method is called, the thread is considered "alive."

    join(): This method blocks the calling thread until the thread whose join() method is called completes. 
    The calling thread is suspended until the called thread completes its task. This is useful when you want to wait 
    for a thread to finish its work before proceeding.

    isAlive(): This method returns True if the thread is currently executing (i.e., its run() method is currently running), 
    and False otherwise. This method can be used to check whether a thread is still running or has finished its work
```

```python
#Q.4  Write a python program to create two threads. Thread one must print the list of squares and thread 
#     two must print the list of cubes
import threading

def print_squares():
    for i in range(1, 11):
        print(f"Square of {i} is {i*i}")

def print_cubes():
    for i in range(1, 11):
        print(f"Cube of {i} is {i*i*i}")

if __name__ == "__main__":
    # Creating two threads
    thread1 = threading.Thread(target=print_squares)
    thread2 = threading.Thread(target=print_cubes)

    # Starting both the threads
    thread1.start()
    thread2.start()

    # Waiting for both threads to complete
    thread1.join()
    thread2.join()

    print("Both threads completed!")
```

```python
Q.5 State advantages and disadvantages of multithreading.
Ans: Multithreading has several advantages and disadvantages. 

                                    Advantages of Multithreading:

    Improved performance: Multithreading allows a program to perform multiple tasks simultaneously, 
    which can lead to better utilization of system resources and faster execution of tasks.

    Responsiveness: Multithreading can improve the responsiveness of an application, as it can allow 
    for certain operations to run in the background while the main thread is free to respond to user input.

    Resource sharing: Threads can share resources such as memory, files, and network connections,
    which can lead to more efficient use of resources.

    Simplified programming: Multithreading can simplify programming by allowing complex tasks to be
    broken down into smaller, more manageable threads.

                                    Disadvantages of Multithreading:

    Increased complexity: Multithreading can add complexity to a program, as it requires careful 
    synchronization and coordination between threads to avoid issues such as deadlocks and race conditions.

    Debugging: Debugging a multithreaded program can be difficult, as issues can arise from interactions 
    between threads that are difficult to reproduce and diagnose.

    Resource contention: Multiple threads accessing the same resources can lead to contention and performance
    issues, as threads may need to wait for access to a shared resource.

    Overhead: Creating and managing threads can have overhead costs in terms of memory usage and CPU time.
```

```python
Q.6 Explain deadlocks and race conditions
Ans: Deadlocks and race conditions are two common synchronization problems that can arise in
    multithreaded programs.

    A deadlock occurs when two or more threads are blocked, each waiting for the other to release a resource
    or take a specific action, resulting in a situation where none of the threads can continue. Deadlocks 
    can occur when there is a circular dependency between threads and resources they require. For example,
    if thread A holds resource X and is waiting for resource Y, and thread B holds resource Y and is waiting 
    for resource X, then a deadlock occurs, and both threads will be stuck indefinitely.

    A race condition occurs when two or more threads access shared resources or variables in a way that can 
    result in unexpected or incorrect behavior. In a race condition, the final outcome depends on the order
    of execution of the threads, which can be unpredictable. For example, if two threads try to increment the
    same variable at the same time, the final value of the variable may not be what was expected due to
    interleaving of the instructions of the two threads.

```
